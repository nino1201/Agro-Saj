```python
import numpy as np
import matplotlib.pyplot as plt
from tkinter import *
from PIL import ImageTk,Image
import sqlite3
import mysql.connector
```

```python
root = Tk()
root.title("La Nola Viniloteca")
root.geometry("400x400")

#Databases

#crea una base de datos
#conn= mysql.connector.connect(
#        host="",
#        user="",
#        passwd="",
#    )

conn= sqlite3.connect("AlejoDiscos1.db")


c=conn.cursor()
def submit():
    
    
    c.execute("INSERT INTO addresses VALUES (:f_name,:Album,:click,:Interprete,:Genero,:Posicion,:Disco )",
             {
                 "f_name": f_name.get(),
                 "Album": Album.get(),
                 "click": clicked.get(),
                 "Interprete": Interprete.get(),
                 "Genero": Genero.get(),
                 "Posicion": Posicion.get(),
                 "Disco": Disco.get(),
             })
    
    
    #commit Changes
    conn.commit()

    #close Conection
    conn.close()
    
    f_name.delete(0,END)
    Album.delete(0,END)
    clicked.set("A")
    Interprete.delete(0,END)
    Genero.delete(0,END)
    Posicion.delete(0,END)
    Disco.delete(0,END)
    
#create table
#c.execute("""CREATE TABLE addresses(
#        Nombre text,
#        Album text,
#        Lado text,
#        Interprete text,
#        Genero text, 
#        Posicion text,
#        Disco integer
#        )""")
f_name=Entry(root, width=30)
f_name.grid(row=0, column=1,padx=20 )

Album=Entry(root, width=30)
Album.grid(row=1, column=1,padx=20 )

clicked=StringVar()
clicked.set("A")
Lado=OptionMenu(root, clicked, "A", "B")
Lado.grid(row=2, column=1,padx=20 )

Interprete=Entry(root, width=30)
Interprete.grid(row=3, column=1,padx=20 )

Genero=Entry(root, width=30,)
Genero.grid(row=4, column=1,padx=20 )

Posicion=Entry(root, width=30)
Posicion.grid(row=5, column=1,padx=20 )

Disco=Entry(root, width=30)
Disco.grid(row=6, column=1,padx=20 )

f_nameLabel= Label(root, text="Nombre: ")
f_nameLabel.grid(row=0, column=0,padx=20 )

f_AlbumLabel= Label(root, text="Album: ")
f_AlbumLabel.grid(row=1, column=0,padx=20 )

f_LadoLabel= Label(root, text="Lado: ")
f_LadoLabel.grid(row=2, column=0,padx=20 )

f_InterpreteLabel= Label(root, text="Interprete: ")
f_InterpreteLabel.grid(row=3, column=0,padx=20 )

f_GeneroLabel= Label(root, text="Genero: ")
f_GeneroLabel.grid(row=4, column=0,padx=20 )

f_PosicionLabel= Label(root, text="Posici√≥n: ")
f_PosicionLabel.grid(row=5, column=0,padx=20 )

f_DiscoLabel= Label(root, text="Disco: ")
f_DiscoLabel.grid(row=6, column=0,padx=20 )

#Create Submit Button
guardar= Button(root, text="Guardar en la Base de Datos", command=submit)
guardar.grid(row=7, column=0, columnspan=2, pady=10,padx=10,ipadx=100 )
#commit Changes
conn.commit()

#close Conection
conn.close()

root.mainloop()#commit Changes
```
